```python
import numpy as n
import pandas as p
import seaborn as sb
from matplotlib import pyplot as plt
```

```python
#Read Excel File
Emp_Dataset = p.read_excel(r"C:\Users\1\Desktop\PB\myexcel.xlsx")
Emp_Dataset
```

```python
# Data Pre Processing Replacing Null values using Random Numbers
Emp_Dataset["Height"]= n.random.randint(150 , 185 ,458)
Emp_Dataset["Weight"]= n.random.randint(60 , 100 ,458)

Emp_Dataset
```

```python
# 1. Calculate total count by team 
Team_Count =  Emp_Dataset["Team"].value_counts()

# How many are there in each Team and the percentage splitting with respect to the total employees.
Team_Name=p.Series(Emp_Dataset["Team"].unique())
print(Team_Count)
# Using Pie Chart
print("\n\n-----PIE CHART------\n\n")
plt.pie(Team_Count ,labels=Team_Name,autopct="%1.1f%%")
plt.figure(figsize= (50, 50))
plt.show()

# BAr chart for plotting Team 
print("\n\n-----BAR CHART-----\n\n")
plt.bar(Team_Name,Team_Count, color = "Green")


plt.figure(figsize= (50,100))
plt.show()
```

```python
Emp_Dataset.to_csv(r"C:\Users\1\Desktop\PB\myexcel_Final.csv")
```

```python
# Total count of employees in different position

Pos_Cnt =  Emp_Dataset["Position"].value_counts()
Pos_Cnt
```

```python
#2.  Segregate the employees w.r.t different positions.
print("****Employees in SG Position*****\n",Emp_Dataset.loc[(Emp_Dataset["Position"]=="SG")])
print("\n****Employees in PF Position****\n",Emp_Dataset.loc[(Emp_Dataset["Position"]=="PF")])
print("\n-----Employees in PG Position-----\n",Emp_Dataset.loc[(Emp_Dataset["Position"]=="PG")])
print("\n++++++Employees in SF Position+++++\n",Emp_Dataset.loc[(Emp_Dataset["Position"]=="SF")])
print("\n*****Employees in C Position******\n",Emp_Dataset.loc[(Emp_Dataset["Position"]=="C")])
```

```python
# 3..Find from which age group most of the employees belong to.
print(Emp_Dataset["Age"].value_counts())
DF=Emp_Dataset[(Emp_Dataset["Age"]>=24) & (Emp_Dataset["Age"]<=36)]
print("\n\n----Details of employees---- \n\n",DF)
```

```python
# Histogram for plotting the Age group where most of the employees belong

age=Emp_Dataset["Age"]
age
plt.hist(age,bins=[15,20,25,30,35,40,45,50,55])
plt.legend("Age")
plt.show()
```

```python
#4. Find out under which team and position, spending in terms of salary is high
s=Emp_Dataset["Salary"].max()
Emp_Dataset[(Emp_Dataset["Salary"]==s)]

```

```python
# Position and count scatter plot
pos=Emp_Dataset["Position"].unique()
plt.scatter(pos, Pos_Cnt, color="violet")
plt.figure(figsize= (50, 50))
plt.show()
```

```python
#5.Find if there is any correlation between age and salary , represent it visually.

New=Emp_Dataset[["Age","Salary"]]
print("Correlation between Age and Salary---\n\n")
print(New.corr())

#Correlation Matrix 

# Correlation Matrix of Age and Salary

print("\n-----Correlation Matrix------ \n\n")
cbar = False
linewidths=1.5
linecolor = "white"
sb.heatmap(New.corr(),linewidth=linewidths , linecolor=linecolor,cbar=cbar, annot=True,cmap="PiYG" )
```
